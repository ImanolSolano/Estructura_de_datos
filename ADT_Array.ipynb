```python
"""
ADT Array
"""
class Array():
  def __init__(self, tamanio):
    self.__tamanio = tamanio
    self.__data = [1 for x in range(self.__tamanio)]
  
  def get_item(self,index):
    return self.__data[index]
  
  def set_item(self,dato):
    self.__dato = dato
    return self.__dato

  def to_string(self):
    print(self.__data)

  def get_length(self):
    print(self.__tamanio)

  def clear(self,dato):
    self.__data.clear()
    print(self.__data)
    self.__data = [dato for x in range(self.__tamanio)]
    return self.__data
```

```python
ejemplo = Array(7)

ejemplo.to_string()

#En el momento que se ocupa el lenght, no imprime el get_item, pero si funciona
ejemplo.get_item(3)

#ejemplo.set_item(6,4)

ejemplo.get_length()

ejemplo.clear(4)
```
